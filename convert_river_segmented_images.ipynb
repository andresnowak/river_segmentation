```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
path = "river_segmented_hydrology/masks"
path_new = "river_segmented_hydrology_v2/masks"
path_for_model = "dataset/dataset_v3/train/masks"
```

```python
images_path = os.listdir(path)
len(images_path)
```

```python
for image_path in images_path:
    image = cv2.imread(os.path.join(path,image_path), cv2.COLOR_BGR2GRAY)
    
    unique_mask = np.unique(image)
    for i in unique_mask:
        if i != 0:
            image[image == i] = 1

    #print(np.unique(image))

    plt.imshow(image)
    plt.show()

    cv2.imwrite(os.path.join(path_new, image_path), image)
    cv2.imwrite(os.path.join(path_for_model, image_path), image)
```
